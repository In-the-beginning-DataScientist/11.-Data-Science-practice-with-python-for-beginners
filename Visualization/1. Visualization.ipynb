## 2. JOINT PLOT
#### Advance version of distribution plot

```python
# Joint Plot is very useful if the Goal is to perform Correlation Analysis w.r.t 
#       1. FeatureCol v/s Label (Checking LinearRelationship. If Strong, preserve else discard feature)
#       2. Featurecol1 v/s Featurecol2 (Multicollinearity. If exists, Remove Else Preserve)
```

```python
sns.jointplot(x = tips.tip , y= tips.total_bill)
```

```python
# The above plot gives linear relationship of feature and label and also skewness of data independantly.
# thrue the scotter plot in the above plot, unable to understand whether its having linear
# relationship or not..So for that we can use the formula called regression in the plot
```

```python
sns.jointplot(x = tips.tip , y= tips.total_bill, kind='reg')
```

```python
# there is a regression line which has drawn with the help of pearson Rformula with the help of data points.
# Without numbers, just by visual graphic representation, 
# understanding the isghts of the data is called Visual EDA
# may not have the best regression line but its a ideal line for the given data.
```

### Inference or Observation

```python
# 1. Most of my customer spend atleast $8 in a single bill.
# Average no of customers might give me the business of $13 per table, 
# THIS WHERE WE CAN see the mean value in the histograph which is $13 (this from right side histogram)
# All the customers an Average amount of $2-$3 paid as tip (from the top Histogram)
```
